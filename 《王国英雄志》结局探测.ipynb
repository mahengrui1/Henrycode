```python
yx=12
se=6
history='从现在开始'
choose=[]#符合语法 无用

def Theoutput(choose):
    global yx, se, history
    
    
    
    i1='，求婚3选了1';i2='，求婚3选了2';i3='，求婚3选了3';i4='，大喊2选了1';i5='，大喊2选了2';i6='，迁都2选了1';i7='，迁都2选了2'
    i8='，骨灰2选了1';i9='，骨灰2选了2';i10='，大王3选了1';i11='，大王3选了2';i12='，大王3选了3'
    
    #print(2,choose)
    
    if len(history)==5:#从现在开始
        history += '：'
        Theinput([[3,-5],[1,2],[-2,3]],'OR')#求婚3
        return
    
    if len(history)==6:
        #print(2,choose)
        if(choose == [3,-5]):
            history += i1
        elif(choose == [1,2]):
            history += i2
        elif(choose == [-2,3]):
            history += i3
        else:
            print('Fail1',choose)
        
        Theinput([[[-3,3],[3,-2]],[[2,-2],[-2,2]]],'MOR')#大喊2 MOR 迁都2
        #Theinput([[2,-2],[-2,2]],'OR')#迁都2
        #print(100)
        return
    
    #print(history[-6:-1],i1[:-1])
    
    
    if history[history.rfind('，'):][:6]==i1[:-1]:
        if(choose == [-3,3]):#大喊2
            history += i4
            #print(100)
            Theinput([[2,-2],[-2,2]],'OR')#激活迁都2
        elif(choose == [3,-2]):
            history += i5
            Theinput([[2,-2],[-2,2]],'OR')#激活迁都2
            
        elif(choose == [2,-2]):#迁都2
            history += i6
            print('检索完毕：')
        elif(choose == [-2,2]):
            print('检索完毕：')
            history += i7
        else:
            print('Fail2',choose)

        return
    
    if history[history.rfind('，'):][:6]==i4[:-1]:
        if(choose == [2,-2]):#大喊后 迁都2
            history += i6
            #print(100)
            #Theinput([[2,-2],[-2,2]],'OR')#激活迁都2
        elif(choose == [-2,2]):
            history += i7            
        else:
            print('Fail3',choose)
            
        Theinput([[-3,3],[2,-1]],'OR')#激活骨灰2
        return
    
    if history[history.rfind('，'):][:6]==i6[:-1]:
        if(choose == [-3,3]):#骨灰2
            history += i8
            #print(100)
            #Theinput([[2,-2],[-2,2]],'OR')#激活迁都2
        elif(choose == [2,-1]):
            history += i9      
        else:
            print('Fail4',choose)
            
        Theinput([[3,-2],[2,3],[-3,-3]],'OR')#激活大王3
        return
    
    if history[history.rfind('，'):][:6]==i8[:-1]:
        print('检索完毕：')
        if(choose == [3,-2]):#大王3
            history += i10
            #print(100)
            #Theinput([[2,-2],[-2,2]],'OR')#激活迁都2
        elif(choose == [2,3]):
            history += i11 
        elif(choose == [-3,-3]):
            history += i12
        else:
            print('Fail5',choose)
            
        #Theinput([[3,-2],[2,3],[-3,3]],'OR')#激活大王3
        return
    
    print('Fail5',choose)
```
